```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r'C:\Users\Chris\Desktop\PFDA\assignments\people-1000.csv'
data = pd.read_csv(file_path)

# Extract domain names
data['Domain'] = data['Email'].apply(lambda x: x.split('@')[-1])

# Count domain occurrences
domain_counts = data['Domain'].value_counts()

# Create the pie chart
colors = ['#ff7043', '#ffa726', '#ffcc80'] 
plt.figure(figsize=(8, 8))
plt.pie(domain_counts, labels=domain_counts.index, autopct='%1.1f%%', startangle=90, colors=colors)

plt.title('Distribution of Email Domains', fontsize=14)
plt.axis('equal')
plt.show()

```
