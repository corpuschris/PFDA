```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 1000 battles
num_battles = 1000
attacker_losses = []
defender_losses = []

for _ in range(num_battles):
    # Roll dice for attacker and defender
    attacker_dice = sorted(np.random.randint(1, 7, 3), reverse=True)
    defender_dice = sorted(np.random.randint(1, 7, 2), reverse=True)

    # Calculate losses
    attacker_loss = 0
    defender_loss = 0
    for i in range(2):
        if attacker_dice[i] <= defender_dice[i]:
            attacker_loss += 1
        else:
            defender_loss += 1

    attacker_losses.append(attacker_loss)
    defender_losses.append(defender_loss)

# Create the histogram
plt.hist([attacker_losses, defender_losses], bins=3, alpha=0.7, color=['orange', 'teal'], label=['Attacker Losses', 'Defender Losses'])

# Add title and axis labels
plt.title('Risk Battle Simulation')
plt.xlabel('Casualties')
plt.ylabel('Frequency')

# Add a grid
plt.grid(True)

# Add the legend below the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)

# Show the plot
plt.show()
```
