```python
#Assignment week 2: plotting temperature

#Author: Christiano Ferreira

# Step 1: Import libraries
import pandas as pd  
import matplotlib.pyplot as plt 

# Step 2: Load the weather data from the CSV file
data = pd.read_csv('weatherreadings1.csv')  

# Step 3: Convert the time column to date format
data['reportStartDateTime'] = pd.to_datetime(data['reportStartDateTime']) 

# Step 4: Create a plot
plt.plot(data['reportStartDateTime'], data['dryBulbTemperature_Celsius'], color='blue') 

# Step 5: Add a title and labels
plt.title('Temperature Over Time')  
plt.xlabel('Time') 
plt.ylabel('Temperature (Â°C)')  

# Step 6: Show the plot
plt.xticks(rotation=45)  
plt.grid(True)  
plt.show()  
```
